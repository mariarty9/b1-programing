```python
###Defining a simple class

class Car:
    wheels = 4 #Class shared by all instances

    def display_info(self):
        print(f"This car has {self.wheels} wheels")

#Creating objects from a class
my_car = Car()
your_car = Car()

my_car.display_info()
your_car.display_info()
```

```python
#Instance Attribute
my_car.colour = "red"

#Class
```
