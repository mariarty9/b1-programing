```python
#EXERCISE 1: Personal Music Library Manager

songs = []
genre_count = {}
print("=== Hello! Welcome to your Music Library Management!===")

#Collect songs from the input

for i in range(1,6):
    print(f"Enter Song: {i}")
    song_name = input("Song name: ")
    genre = input("Genre: ")

#Use tuples to store songs
    
    tuple_songs = (song_name, genre)
    songs.append(tuple_songs)

#Count genres in the dictionary

    genre_count[genre] = genre_count.get(genre, 0) + 1

#Display music library & genre statistocs
print("")
print("======== YOUR MUSIC LIBRARY ========")
for index, (song, genre) in enumerate(songs, 1): # enumerate is a built-in function. It adds a counter to an iterable (like a sequence) and returns it as an enumerate object.
    print(f"{index}. {song} ({genre})")

print("")
print("========= GENRE STATISTICS =========")

for genre, count in genre_count.items():
    print(f"{genre.capitalize()}: {count} songs")


#Establish the most popular genre
popular_genre = max(genre_count, key = genre_count.get)
print("")
print(f"Most popular genre: {popular_genre}")
```

```python
#EXERCISE 2:Student Grade Analyzer

student_records = []
stats = {}

print("======== Grade Analyzer ========")
for i in range(1,7):
    student_name = input(f"Enter the name of the {i} student: ")
    grade = int(input(f"Student {i} received: "))
    student_records.append((student_name, grade))


#Calculate stats
grades = [grade for student_name, grade in student_records]
stats['highest'] = max(grades)
stats['lowest'] = min(grades)
stats['average'] = sum(grades)/len(grades)

#Unique grades
unique_grades = set(grades)

grade_distribution = {}
for grade in grades:
    grade_distribution[grade] = grade_distribution.get(grade, 0) + 1

#Display of the analyzer
print("")
print("======== STUDENT RECORDS ========")

for index, (student_name, grade) in enumerate(student_records, 1):
    print(f"{index}. {student_name}: ({grade})")

print("")
print("======== CLASSIC STATISTICS ========")
print(f"Highest grade is: {stats['highest']}")
print(f"Lowest grade is: {stats['lowest']}")
print(f"Average grade is: {stats['average']}")


print("")
print("======== UNIQUE GRADES ========")
print(unique_grades)
print(f"Total unique grades: {len(unique_grades)}")

print("")
print("======== GRADE DISTRIBUTION ========")

for grade in sorted(grade_distribution.keys(), reverse = True):
    count = grade_distribution[grade]
    plural = "students" if count != 1 else "student"
    print(f"{grade}: {count} {plural}")
    
```

```python
#EXERCISE 3: Personal Expense Tracker

expense_records = [] 
category_totals = {} 
unique_categories = set()

print("======== PERSONAL EXPENSE TRACKER ========")

num_expenses = 5

for i in range(1,num_expenses + 1):
    
    category = input(f"Enter expense {i} category:")
    amount = float(input(f"Enter expense {i} amount:"))
    date = input(f"Enter expense {i} date (YYYY-MM-DD):")
    expense_records.append((category, amount, date))

# Category and sum of expenses
for category, amount, date in expense_records:
    unique_categories.add(category)
    category_totals[category] = category_totals.get(category, 0) + amount

#Calculating statistics
total_spending = 0
highest_expense = None
lowest_expense = None

for category, amount, date in expense_records:
    total_spending += amount

    if highest_expense is None or amount > highest_expense[1]:
        highest_expense = (category, amount, date)

    if lowest_expense is None or amount < lowest_expense[1]:
        lowest_expense = (category, amount, date)

average_expense = total_spending / len(expense_records)

#Personal expenses logs
print("\n======== OVERALL SPENDING SUMMARY ========")
print(f"Total Spending: ${total_spending:.2f}")
print(f"Average Expense: ${average_expense:.2f}")

#Highest expense
highest_category = highest_expense[0]
highest_amount = highest_expense[1]
highest_date = highest_expense[2]

print(f"Highest Expense: ${highest_amount:.2f} (Category: {highest_category}, Date: {highest_date})")

#Lowest expenses
lowest_category = lowest_expense[0]
lowest_amount = lowest_expense[1]
lowest_date = lowest_expense[2]

print(f"Lowest Expense: ${lowest_amount:.2f} (Category: {lowest_category}, Date: {lowest_date})")

print("\n======== UNIQUE CATEGORIES SPENT ON ========")
print(unique_categories)
print(f"Total unique categories: {len(unique_categories)}")

print("\n======== SPENDING BY CATEGORY ========")
for category, total in category_totals.items():
    print(f"{category}: ${total:.2f}")
```
