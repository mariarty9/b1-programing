```python
### Practical example: word count

sentence = "the cat and the dog and the bird"
words = sentence.split()

# Create empty dictionary
word_count = {}

# Count each word
for word in words:
 if word in word_count:
 word_count[word] = word_count[word] + 1
 else:
 word_count[word] = 1

# Display results
print("Word frequencies:")
for word, count in word_count.items():
 print(f"'{word}': {count}")

'''How It Works
Start with empty dictionary1.
For each word, check if it exists2.
If yes, increment count3.
If no, set count to 14.
Display final counts''' !
```

```python
###SOLVIG THE DATA TRANSFORMATION PROBLEM
# Sample raw data
raw_data = """alice,success
bob,failed
alice,failed
diana,success
"""

# Split into lines
lines = raw_data.strip().split('\n')
print(f"Lines: {lines}")

# Example of splitting a single line
first_line_parts = lines[0].split(',')
print(f"First line parts: {first_line_parts}")

#Create a new dictionary
user_attempts = {}
print(f"Initial dictionary: {user_attempts}")

# Process each line
for line in lines:
    username, status = line.split(',') # status is not used for counting attempts, but parsed
    user_attempts[username] = user_attempts.get(username, 0) + 1 # Increment count

print(f"Final user attempts: {user_attempts}")
```

```python
###FUNCTION & DICTIONARIES
def summarize_attempts(attempts):
 """Count total login attempts per user"""
 summary = {}

for user, result in attempts:
 # If user exists, increment count
 # If user is new, start at 1
    if user in summary:
        summary[user] = summary[user] + 1
    else:
        summary[user] = 1
return summary
    
# Our data
login_attempts = [
 ("alice", "success"),
 ("bob", "failed"),
 ("alice", "failed"),
 ("dana", "success"),
 ("bob", "success"),
 ("alice", "success")
]
# Call the function
result = summarize_attempts(login_attempts)
print(result)
```

```python
from collections import Counter

# Count login attempts
login_attempts = ["alice", "bob", "alice", "charlie", "alice", "bob", "diana"]
counts = Counter(login_attempts)

# Most common users
print("Top 2 users:")
print(counts.most_common(2))
# [('alice', 3), ('bob', 2)]

# Total count
print("Total attempts:", sum(counts.values()))

# Add more data
more_logins = ["alice", "emma", "charlie"]
counts.update(more_logins)
print("Updated counts:", counts)

# Find users with exactly 2 attempts
for user, count in counts.items():
 if count == 2:
     print(f"{user} logged in exactly twice")
```

```python
#DEMO: SAVING A DICTIONARY TO JSON
import json
# Our data
user_scores = {
 "alice": 3,
 "bob": 5,
 "charlie": 2,
 "diana": 7
}
# Save to file
with open("scores.json", "w") as file:
 json.dump(user_scores, file, indent=4)
print("Data saved to scores.json")
# What the file contains:
# {
# "alice": 3,
# "bob": 5,
# "charlie": 2,
# "diana": 7
# }
```

```python
import json
# Load from file
with open("scores.json", "r") as file:
 loaded_scores = json.load(file)
print("Loaded data:", loaded_scores)
print("Type:", type(loaded_scores))
# Use the data normally
for user, score in loaded_scores.items():
 print(f"{user}: {score}")
# Modify and save again
loaded_scores["emma"] = 4
with open("scores.json", "w") as file:
 json.dump(loaded_scores, file, indent=4)

print("Updated data saved!")
```
