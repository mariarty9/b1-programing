```python
#Exercise 1: Failed Login Detector
import time
from colorama import Fore, Back, Style 

login_attempts = [("Vi", "success"), ("Jayce", "failed"), ("Jayce", "failed"), ("Caitlyn", "success"), ("Jayce", "failed"), ("Vi", "failed"), ("Jinx", "failed"), ("Jinx", "failed"), ("Jinx", "failed")]

print("Checking login attempts...")
time.sleep(3)

failure = {}

for attempt in login_attempts:
    user, status = attempt
    if status == "failed":
        if user in failure:
            failure[user] += 1
        else:
            failure[user] = 1

for user in failure:
    if failure[user] >= 3:
        print(Fore.RED + f"ALERT: User '{user}' has {failure[user]} failed login attempts")

print(Fore.GREEN + "Security check complete")
```

```python
# Exercise 2: Port Security Scanner
devices = [ ("192.168.1.10", [22, 80, 443]), ("192.168.1.11", [21, 22, 80]), ("192.168.1.12", [23, 80, 3389])]
risky_ports = [21, 23, 3389]
risks_found = 0

print("Scanning network devices... ")
for device in devices:
    ip, open_ports = device
    for port in open_ports:
        if port in risky_ports:
            print(f"WARNING: {ip} has risky port {port} open")
            risks_found += 1
print(f"Scan complete: {risks_found} security risks found")
    
```

```python
# Excercise 3: Password Policy Validator
passwords = [ "Pass123", "SecurePassword1", "weak", "MyP@ssw0rd", "NOLOWER123"]

print("Validating passwords...")

compliant = 0
non_compliant = 0
min_length = 8
for password in passwords:
    errors = []

# Check length of the password
    if len(password) < min_length:
        errors.append("Too short")

# Flags
    with_upper = False
    with_lower = False
    with_digit = False

# Check characters in password one by one
    for ch in password:
        if 'A' <= ch <= 'Z':
            with_upper = True
        elif 'a' <= ch <= 'z':
            with_lower = True
        elif '0' <= ch <= '9':
            with_digit = True

# If the requirements don't meet the policy
    if not with_upper:
        errors.append("No uppercase letter")
    if not with_lower:
        errors.append("No lowercase letter")
    if not with_digit:
        errors.append("No digit")

# Output of the results
    if len(errors) == 0:
        print(f"PASS: '{password}' - Meets all requirements")
        compliant += 1
    else:
        print(f"FAIL: '{password}' - {', '.join(errors)}")
        non_compliant += 1

# Summary
print(f"Summary: {compliant} compliant, {non_compliant} non-compliant")
```
