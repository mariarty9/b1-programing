```python
# Simple password Strength Checker

password = "MyP@ssw0rd"
score = 0

for char in password:
    if char.isupper():
        score += 1
    elif char.islower():
        score += 1
    elif char.isdigit():
        score += 1

print(f"Password strength: {score}/3")
```
