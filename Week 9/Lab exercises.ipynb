```python
#Lab Task 1: Building a school management system

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"Greetings, my name is {self.name} and I'm {self.age} years old."

class Student(Person):  #Creating the inheritance from the class Person
    def __init__(self, name, age, student_id):
        super().__init__(name, age) #Accesig the constructor of the Parent Class
        self.student_id = student_id
        
    def introduce(self):
        return f"Greetings, I'm {self.name} and I'm {self.age} years old. My student id is {self.student_id}. "

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def introduce(self):
        return f"Greeting, my name is {self.name} and I am your new teacher. I'm {self.age} and I'll be teaching {self.subject}"

print("\n-------- SCHOOL MANAGEMENT SYSTEM --------")
print("\n")
student_1 = Student("Alice", 16, "S001")
print(student_1.introduce())
teacher_1 = Teacher("Mr. Smith", 35, "Mathematics")
print(teacher_1.introduce())
```

```python
#Lab Task 2: Library System with Composition 

class Book():
    
    def __init__(self, title, author, ISBN):
        self.title = title
        self.author = author
        self.ISBN = ISBN

    def book_info(self):
        return f"{self.title} by {self.author}, ISBN: {self.ISBN}."

class Library():

    def __init__(self, library_id):
        self.library_id = library_id
        self.books = []
        
    def add_book(self, book):
        self.books.append(book)
        return f"{book.book_info()} is added to the list. "

    def remove_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                self.books.remove(book)
                return f"{book.book_info()} is removed from the list."
        return f"Book '{title}' not found."

    def list_all_books(self):
        if not self.books:
            return f"{self.library_id} has no books"
            
        result = f"\n ======= Books in {self.library_id} ======= \n"
        for i, book in enumerate(self.books, 1):
            result += f"{i} {book.book_info()} \n"
        return result

    def search_by_title(self, search_term):
        found = [book for book in self.books
                if search_term.lower() in book.title.lower()]
        if found:
           result = f"\nFound {len(found)} book(s):\n"
        for book in found:
                result += f"- {book.book_info()}\n"
        return result
        return f"No books found matching '{search_term}'"

library = Library("N213")

book_1 = Book(f"'Fourth Wing'", "Rebecca Yarros", "DE44510200004115249524")
book_2 = Book("'Hunger Games'", "Suzanne Collins", "DE98300205003248891485")
book_3 = Book("'The Circle'", "Dave Eggers", "DE42500202006264469889")

print(library.add_book(book_1))
print(library.add_book(book_2))
print(library.add_book(book_3))

#Listing of the books
print(library.list_all_books())

#Searching for the book by its title
print(library.search_by_title("hunger games"))

#Removing the book from the list
print(library.remove_book("The Circle"))
```

```python
#Lab Task 3: File Manager with os Module

import os

def demo():

    #Gertting the current working directory
    cwd = os.getcwd()
    print("Current working directory:", cwd)

    #Creating new folder
    new_folder = "lab_files"
    if not os.path.exists(new_folder):
        os.mkdir(new_folder)
        print(f"New folder created: {new_folder}")
    else:
        print(f"Folder '{new_folder}' already exists")

    #Creating three empty text files
    file_names = ["file1.txt", "file2.txt", "file3.txt"]
    for file_name in file_names:
        file_path = os.path.join(new_folder, file_name)
        with open(file_path, 'w') as f:
            f.write(f"This is {file_name}")
        print(f"Created file: {file_name}")

    #Listing all files in the folder
    
    print(f"\n ======== Files in {new_folder} ========")
    files = os.listdir(new_folder)
    for file in files:
        print(f"- {file}")
        
    #Renaming one of the files
    old_path = os.path.join(new_folder, "file2.txt")
    new_path = os.path.join(new_folder, "renamed_file.txt")
    if os.path.exists(old_path):
        os.rename(old_path, new_path)
        print(f"Renamed 'file2.txt' to 'renamed_file.txt'")


    #Displaying the files after rename
    print(f"\n ======== Files after rename ========")
    files = os.listdir(new_folder)
    for file in files:
        print(f"-{file}")

    #Cleaning up/Removing all the files and the folder
    print(f"\n ======== Clean Up ========")
    for file in os.listdir(new_folder):
        file_path = os.path.join(new_folder, file)
        os.remove(file_path)
        print(f"Deleted: {file}")


    os.rmdir(new_folder)
    print(f"Removed folder: {new_folder}")
    print(f"The clean up was successfully completed.")

#Running the method
demo()
```
