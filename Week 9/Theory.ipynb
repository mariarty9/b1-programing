```python
# Building a class

class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_checked_out = False
        
    def display_info(self):
        status = "available" if not self.is_checked_out else "checked out"
        return f"'{self.title}' by {self.author} (ISBN: {self.isbn}) - Status: {status}"
    
    def check_out(self):
        if not self.is_checked_out:
            self.is_checked_out = True
            return f"'{self.title}' has been checked out."
        else:
            return f"'{self.title}' is already checked out."

# Creating a Book object
my_book = Book("The Great Adventure", "Jane Doe", "978-1234567890")

# Calling a method on the object
print(my_book.display_info())
print(my_book.check_out())
print(my_book.display_info())
```

```python
# Basic Inheritance Syntax
class ParentClass:
    def __init__(self, name):
        self.name = name
            
    def greet(self):
        return f"Hello, I'm {self.name}"
        
class ChildClass(ParentClass):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

```

```python
### Parent class animal
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return "Some sound"
    def info(self):
        return f"I am {self.name}"
        
# Example Usage for Animal
print("--- Parent Class: Animal ---")
animal_obj = Animal("Generic Animal")
print(f"Animal Name: {animal_obj.name}")
print(f"Animal Speak: {animal_obj.speak()}")
print(f"Animal Info: {animal_obj.info()}")

### Child class Dog
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
    def speak(self):
        return "Woof!"
        
# Example Usage for Dog
print("\n--- Child Class: Dog ---")
dog_obj = Dog("Buddy", "Golden Retriever")
print(f"Dog Name: {dog_obj.name}")
print(f"Dog Breed: {dog_obj.breed}")
print(f"Dog Speak (overridden): {dog_obj.speak()}")
print(f"Dog Info (inherited): {dog_obj.info()}")
```

```python
###Multi level inheritance

class LivingThing:
    def breathe(self):
        return "Breathing..."
    
class Animal(LivingThing):
    def move(self):
        return "Moving..."
    
class Dog(Animal):
    def bark(self):
        return "Woof!"
    
buddy = Dog()
print(buddy.breathe()) # From LivingThing
print(buddy.move()) # From Animal
print(buddy.bark()) # From Dog
```

```python
### Syntax for overriding 

class Parent:
    def greet(self):
        return "Hello from Parent"
        
class Child(Parent):
    def greet(self):   #Child calss verrides the Parent class
        return "Hello from Child"
        
obj = Child()
print(obj.greet()) # Output: Hello from Child
```

```python
### Using super() Explained
class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        parent_sound = super().speak()
        return f"{parent_sound} like Woof!"

dog = Dog()
print(dog.speak()) 
```

```python
### Overriding __str__
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, aged {self.age}"

person = Person("Alice", 30)
print(person)
```

```python
###Composition

class Engine:
    def start(self):
        return "Engine started"
    def stop(self):
        return "Engine stopped"
        
class Car:
    def __init__(self, brand):
        self.brand = brand
        self.engine = Engine() # Composition
    def start_car(self):
        return f"{self.brand}: {self.engine.start()}"
        
my_car = Car("Toyota")
print(my_car.start_car()) # Toyota: Engine started
```

```python
class Customer:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        
    def get_contact(self):
        return f"{self.name} ({self.email})"
        
class Order:
    def __init__(self, order_id, customer):
        self.order_id = order_id
        self.customer = customer # Composition: an Order *has a* Customer
        
    def summary(self):
        return f"Order {self.order_id} for {self.customer.get_contact()}"
        
# Example usage
customer = Customer("Bob", "bob@example.com")
order = Order(12345, customer)
print(order.summary())
```

```python
###Formating dates
from datetime import datetime

now = datetime.now()

# Different format examples
print(now.strftime("%Y-%m-%d")) # 2024-01-15
print(now.strftime("%d/%m/%Y")) # 15/01/2024
print(now.strftime("%B %d, %Y")) # January 15, 2024
print(now.strftime("%I:%M %p")) # 02:30 PM
print(now.strftime("%A, %B %d, %Y")) # Monday, January 15, 2024
```

```python
#Adding days and time
from datetime import datetime, timedelta
today = datetime.now()
print(f"Today: {today.strftime('%Y-%m-%d')}")

# Add 7 days
next_week = today + timedelta(days=7)
print(f"Next week: {next_week.strftime('%Y-%m-%d')}")

# Subtract 30 days
last_month = today - timedelta(days=30)
print(f"30 days ago: {last_month.strftime('%Y-%m-%d')}")

# Add hours and minutes
meeting_time = today + timedelta(hours=2, minutes=30)
print(f"Meeting in 2.5 hours: {meeting_time.strftime('%I:%M %p')}")
```
